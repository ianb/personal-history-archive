```python
%autoreload
import pha
from pha import strip_url_to_pattern
archive = pha.Archive.default_location()
print(archive)
```

```python
histories = archive.sample_histories_with_page(10, unique_url=True, unique_domain=False)
print("Histories:", len(histories), "Domains:", len(set(h.domain for h in histories)))
print("Patterns:", [strip_url_to_pattern(h.url) for h in histories])
```

```python
%autoreload
from pha import summarytools

for h in histories:
    summary_readable = h.page.readable_text and summarytools.textteaser_summary(h.page)
    summary = summarytools.textteaser_summary(h.page, try_readable=False)
    h.page.display_page(readable=True)
    if summary_readable:
        print(
            "Readable summary:", 
            summarytools.normalize_sentences(summary_readable))
        print()
    print("Normal summary:", summarytools.normalize_sentences(summary))
```
